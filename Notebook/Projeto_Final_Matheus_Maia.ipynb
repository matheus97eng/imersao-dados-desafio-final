#### 2.1) Influência do tempo

Vamos verificar a influência do tempo de exposição nas expressões genéticas, novamente, somente para teste de controle. Para isso, é melhor usar o boxplot para comparar a distribuição para cada tempo de exposição, 24h, 48h ou 72h.

Na "caixa" central, o boxplot irá mostrar aonde a maior parte das expressões genéticas estarão. Os pontos fora dessa "caixa" são os outliers, valores que estão bem distantes da média.

```python
# definindo uma função que calcula as médias para o caso de interesse
def g_por_tempo(tempo, df):
  media = df.query('droga == "controle" and tempo ==tempo').loc[:,'g0':'g771'].mean()
  tempo = str(tempo) + 'h'
  media = pd.concat([media, pd.Series(772*[tempo], index=media.index)], axis=1)   #preenchendo uma coluna com os dados de tempo de exposição
  return media

plt.figure(figsize=(8,8))
data = pd.concat([g_por_tempo(24, df), g_por_tempo(48, df), g_por_tempo(72, df)], axis=0)   #juntando os dados de diferentes tempo em um único df 
data.columns = ['media', 'tempo']   #nomeando colunas
ax = sns.boxplot(x='tempo', y='media', data=data)   #plotando os boxplots    
ax.set_title("média das expressões genéticas para testes de controle", fontsize=15)
ax.set_xlabel('tempo', fontsize=10)
ax.set_ylabel('distribuição das médias\ndas expressões genéticas', fontsize=10)
plt.show()
```

Percebemos pouca influência do tempo nesse caso. Isso é esperado, pois os testes de controle não usam drogas para mudar as expressões genéticas. Mas vejamos a influência do tempo nos testes com a droga mais usada, de código 87d714366. O código python é semelhante ao que fizemos até aqui.

```python
def g_por_tempo(tempo, df):
  media = df.query('droga == "87d714366" and tempo ==tempo').loc[:,'g0':'g771'].mean()
  tempo = str(tempo) + 'h'
  media = pd.concat([media, pd.Series(772*[tempo], index=media.index)], axis=1)
  return media

plt.figure(figsize=(8,8))
data = pd.concat([g_por_tempo(24, df), g_por_tempo(48, df), g_por_tempo(72, df)], axis=0)
data.columns = ['media', 'tempo']
ax = sns.boxplot(x='tempo', y='media',data=data)
ax.set_title('médias das expressões genéticas para testes com a droga "87d714366"', fontsize=15)
ax.set_xlabel('tempo', fontsize=10)
ax.set_ylabel('distribuição das médias\ndas expressões genéticas', fontsize=10)
plt.show()
```

Em um primeiro momento, parece que os gráficos dos testes de controle e os utilizando a droga 87d714366 são os mesmos, todos eles não possuem alterações aparentes com o tempo. Isso significa que a droga 87d714366 não causa maiores efeitos com o passar do tempo. 

Mas, observe que a escala do eixo y de distribuições das médias é muito maior no gráfico para a droga 87d714366. Isso significa que as médias das expressões genéticas variaram muito mais, mesmo que a maioria das médias se encontrem mais perto do zero.
